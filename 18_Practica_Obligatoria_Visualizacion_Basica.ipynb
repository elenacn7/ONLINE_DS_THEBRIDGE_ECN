### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
ciudades_mas_pobladas = df_cali.loc[df_cali.population_total.nlargest(5).index, 
                                    ["city","population_total"]]
```

```python
ciudades_mas_pobladas
```

```python
def separa_nombre(nombre):
    nuevo = nombre[0]
    for letra in nombre[1:]:
        if letra == letra.upper():
            nuevo += " "
        nuevo += letra
    return nuevo
```

```python
ciudades = [ separa_nombre(ciudad) for ciudad in ciudades_mas_pobladas["city"].to_list()]
```

```python
colores = ["blue","gray","red","olive","brown"]
```

```python
plt.figure(figsize = (10,5))
plt.bar(x= ciudades, height= ciudades_mas_pobladas["population_total"]/1000, color = colores, alpha = 0.5)
plt.ylim([0,5000])
plt.ylabel("Habitantes (miles)") #suele ser el eje vertical 
plt.xlabel("Ciudades") #suele ser el eje vertical
```

```python
ciudades_mas_pobladas.set_index("city", inplace = True) # va tomar la columna city como indice
for indice, ciudad in enumerate(ciudades_mas_pobladas.index):  # y para poner un mensaje o texto pues ponemos .text
    poblacion = ciudades_mas_pobladas.loc[ciudad,"population_total"]/1000
    plt.text(ciudades[indice], poblacion + 50, f"{poblacion}", ha = "center") #ha = right o left dependiendo de donde quiera colocarla. 
plt.title("Ciudades ms pobladas de California") # y al ejecutarlo 
plt.show()
```

```python
plt.hist(df_cali["area_total_km2"], color = "steelblue", alpha = 0.7, bins = 50,label = "Hist Area")
plt.xlabel("Superficie ($km^2$)", fontdict = dict(family= "Bradley Hand ITC", size = 18))
plt.ylabel("Nmero de Ciudades", fontdict= dict(family = "Bradley Hand ITC", size = 18))
plt.legend();
```

```python
plt.scatter(df_cali["area_water_km2"], df_cali["population_total"]/1000, alpha = 0.5, color = "green") #scatter es la dispersion
plt.xlabel("rea de agua en la ciudad ($km^2$)")
plt.ylabel("Poblacin (miles habitantes)")
plt.title("Poblacin por kilometro cuadrado de agua")
outlier_1_y = ciudades_mas_pobladas.population_total.max()
outlier_2_y = ciudades_mas_pobladas.population_total.nlargest(2).iloc[1]#poblacion maxima y la segunda mas maxima 
outlier_1_x = df_cali.loc[df_cali.population_total == outlier_1_y,"area_water_km2"].values[0]
outlier_2_x = df_cali.loc[df_cali.population_total == outlier_2_y,"area_water_km2"].values[0]
plt.annotate(xy= (outlier_1_x,outlier_1_y/1000), xytext= (60,3500), text = "Outliers",
              arrowprops= dict(arrowstyle = "->"))#el anotate es las cordenadas que nos esta marcando abajo 
plt.annotate(xy= (outlier_2_x,outlier_2_y/1000), xytext= (60,3500), text = "Outliers",
              arrowprops= dict(arrowstyle = "<->"))#esto es para poder pintarlo el 60 y el 3500
```

```python
plt.boxplot(df_cali["elevation_m"].dropna(), vert = False, boxprops = dict(facecolor = "red"),
               medianprops = dict(color= "black"), patch_artist= True )
plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.yticks([])
```
